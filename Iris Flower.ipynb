```python
# IMPORTING PACKAGES
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sk
import seaborn as sns
```

```python
# IMPORTING IRIS DATASET
from sklearn.datasets import load_iris
data = pd.read_csv('Iris.csv')
```

```python
#LABELLING ACCORDING TO COLOUR

sns.FacetGrid(data,hue="Species", height =6).map(plt.scatter,"SepalLengthCm","PetalLengthCm").add_legend()
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(data,test_size=0.25)
```

```python
train_x = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
train_y =train.Species
```

```python
train_y
```

```python
test_x = test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
test_y =test.Species
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

```

```python
from sklearn import svm
data = svm.SVC()
```

```python
data.fit(train_x,train_y)
pred=data.predict(test_x)
metrics.accuracy_score(pred,test_y)
```

```python
print(classification_report(test_y,pred))
```
